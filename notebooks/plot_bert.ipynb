## MNLI

```python
from plot_mlm import load_data, make_ds_visualize

task = 'mnli'
ds, preds, mlm_out = load_data(task, min_phrase=False, ckpt=None, synonym=False)
```

```python
ds_visualize = make_ds_visualize(ds, preds, mlm_out, task)
```

```python
import plotly.graph_objects as go
import plotly.express as px
import random
import scipy.stats

subset = ds_visualize.shuffle(seed=329).select(list(range(2000)))

# subset = ds_visualize.filter(lambda x: x['color_tag'] == 'NP')
keyx = 'l1d_pred'
keyy = 'l1d_mlm_phrase'


corr, p = scipy.stats.pearsonr(
    ds_visualize[keyx],
    ds_visualize[keyy]
)

xs = [t / 2 for t in subset[keyx]]
ys = [t / 2 for t in subset[keyy]]
# fig = px.scatter(    
#     subset,
#     x=keyx,
#     y=keyy,
#     # color='color_tag',
#     hover_data=['replaced', 'replacement'],
#     # title=f'Pearson r = {corr:.2f}'
# )

fig = go.Figure()
fig.add_trace(
    go.Scatter(x=xs, y=ys, mode='markers', marker_opacity=0.2)
)

fig.update_layout(
    legend=dict(
        orientation='h',
        yanchor="top",
        y=0.99,
        xanchor="center",
        x=0.5
    ),
# margin=dict(l=20, r=0, t=0, b=0),
#     xaxis_title="Step",
    # yaxis_title="$D_{f_0}$",
    xaxis=dict(side='bottom', dtick=0.25),
    yaxis=dict(dtick=0.25),
    font=dict(
        size=18
    ),
    width=500,
    height=500,
    margin=dict(l=0, r=0, t=0, b=0),
)
config = {
  'toImageButtonOptions': {
    'format': 'svg', # one of png, svg, jpeg, webp
    'scale': 1
  }
}
fig.show(config=config)

```
